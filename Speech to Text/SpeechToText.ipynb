Our next work is to import the necessary packages for speech recognition.
In the second line of the below cell we are importing Audio to display our audio file in the google colab. It is same as displaying images rather Image is replayed with Audio in this case.

```
import speech_recognition as sr
from IPython.display import Audio
```

We are operating virtually so we can't directly import the files from our local machine. So, in order to access our audio file we need to first upload our files to the google drive. After that we can mount drive(connect google drive with google colab) and easily access our data from the google colab. I have assigned two variable as speech1 and speech2 for my two audio files.

```
speech1 = "/content/drive/My Drive/Voice/voice1.wav"
speech2 = "/content/drive/My Drive/Voice/voice2.wav"
```

Let's look what is inside our first audio file ie.Speech1

```
Audio(speech1)
```

Initialize our recognizer package using the below one line of code. 

```
init_rec = sr.Recognizer()
```

Now it's time to go through the speech and find our what is inside it. First we need to assign our speech to the AudioFile and send it to the initialized recognizer package. Now we use the google speech recognition to directly convet our speech to the text and printed our speech in the form of text.

The system is not giving result quite accurately but it can't be neglected as an error. The more we try with the different sample it seems to give a better accuracy. However to make it more accurate we have to use the deep learning techniques.

```
with sr.AudioFile(speech1) as speeches:
  audio_data  = init_rec.record(speeches)
  text = init_rec.recognize_google(audio_data)
  print(text)
```

Let's try with the another speech. I have already assigned the speech to variable named Speech2. Repeating the same steps as above for the below lines of code. We can also do everything under a single loop but to understand everything better i have written them separately.

```
Audio(speech2)
```

Trying same process as above but using the different audio sample. In this case it perform slightly better.

```
with sr.AudioFile(speech2) as speeches:
  audio_data  = init_rec.record(speeches)
  text = init_rec.recognize_google(audio_data)
  print(text)
```
